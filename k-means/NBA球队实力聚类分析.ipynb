```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd      # pandas：numpy自带高级函数
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('nba.csv')   # 加载数据
data.head(10)                     # 查看数据的结果
```

```python
minmax_scaler = MinMaxScaler()
# 归一化数据，标准化    # 对数据差距大的进行归一化和标准化，根据具体实例选择性进行归标
# 
X = minmax_scaler.fit_transform(data.iloc[:,1:])     # iloc根据索引获取数据
X[:5]  
```

```python
# 肘部法则
loss = []
for i in range(2,10):
    model = KMeans(n_clusters=i).fit(X)
    loss.append(model.inertia_)
    
plt.plot(range(2,10),loss)
plt.xlabel('k')
plt.ylabel('loss')
plt.show()
```

```python
k = 4
model = KMeans(n_clusters=k).fit(X)
# 将标签整合到原始数据上
data['clusters'] = model.labels_     # 将模型的标签赋到data数据的新列clusters上
data.head()    
```

```python
for i in range(k): 
    print('clusters:',i)
    label_data = data[data['clusters'] == i].iloc[:,0]
    print(label_data.values)
```
